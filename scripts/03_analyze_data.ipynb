## Notebook 03: Data Analysis

### Objective
This notebook focuses on exploratory data analysis (EDA) to understand the relationships between variables and identify patterns that influence term deposit subscription.

### utcomes
- Analyze target variable distribution
- Examine relationships between features and target
- Identify key drivers of term deposit subscription
- Generate statistical summaries for insights

### Steps Covered
1. Load cleaned dataset
2. Analyze target variable distribution
3. Analyze numerical features vs target
4. Analyze categorical features vs target
5. Correlation analysis
6. Generate analysis summary report

### Importing libraries and loading cleaned dataset

```python
### Importing libraries and loading cleaned dataset

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

CLEANED_DATA_PATH = "../outputs/cleaned_data/bank_cleaned.csv"

print("-"*140)
print("LOADING CLEANED DATASET")
print("-"*140)

df = pd.read_csv(CLEANED_DATA_PATH)

print(f"Dataset shape: {df.shape[0]} rows and {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")
print(f"\nFirst 5 rows:")
print(df.head())
```

### Analyzing target variable distribution

```python
### Analyzing target variable distribution

print("-"*140)
print("TARGET VARIABLE DISTRIBUTION")
print("-"*140)

target_counts = df['y'].value_counts()
target_percent = df['y'].value_counts(normalize=True) * 100

print("Target distribution:")
for label in ['no', 'yes']:
    count = target_counts.get(label, 0)
    percent = target_percent.get(label, 0)
    print(f"  {label}: {count} ({percent:.2f}%)")

print(f"\nImbalance ratio (no/yes): {target_counts['no'] / target_counts['yes']:.2f}")

print("\n" + "-"*140)
print("TARGET DISTRIBUTION BY KEY CATEGORICAL VARIABLES")
print("-"*140)

key_categorical = ['job', 'marital', 'education', 'housing', 'loan']

for col in key_categorical:
    print(f"\n{col.upper()}:")
    subscription_rate = df.groupby(col)['y_binary'].mean() * 100
    print(subscription_rate.sort_values(ascending=False).head(3).to_string())
```

### Analyzing numerical features against target

```python
### Analyzing numerical features against target

print("-"*140)
print("NUMERICAL FEATURES ANALYSIS BY TARGET")
print("-"*140)

numeric_cols = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

summary_data = []

for col in numeric_cols:
    mean_no = df[df['y_binary'] == 0][col].mean()
    mean_yes = df[df['y_binary'] == 1][col].mean()
    median_no = df[df['y_binary'] == 0][col].median()
    median_yes = df[df['y_binary'] == 1][col].median()
    diff_percent = ((mean_yes - mean_no) / mean_no) * 100
    
    summary_data.append({
        'Feature': col,
        'Mean (No)': round(mean_no, 2),
        'Mean (Yes)': round(mean_yes, 2),
        'Median (No)': round(median_no, 2),
        'Median (Yes)': round(median_yes, 2),
        'Difference %': round(diff_percent, 2)
    })
    
    print(f"\n{col.upper()}:")
    print(f"  Mean - No: {mean_no:.2f}, Yes: {mean_yes:.2f}")
    print(f"  Median - No: {median_no:.2f}, Yes: {median_yes:.2f}")
    print(f"  Difference: {diff_percent:.2f}%")

print("\n" + "-"*140)
print("NUMERICAL FEATURES SUMMARY TABLE")
print("-"*140)
summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string())
```

### Analyzing categorical features against target

```python
### Analyzing categorical features against target

print("-"*140)
print("CATEGORICAL FEATURES ANALYSIS BY TARGET")
print("-"*140)

categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

categorical_summary = []

for col in categorical_cols:
    subscription_rate = df.groupby(col)['y_binary'].mean() * 100
    max_cat = subscription_rate.idxmax()
    max_rate = subscription_rate.max()
    min_cat = subscription_rate.idxmin()
    min_rate = subscription_rate.min()
    
    categorical_summary.append({
        'Feature': col,
        'Best Category': max_cat,
        'Subscription Rate %': round(max_rate, 2),
        'Worst Category': min_cat,
        'Lowest Rate %': round(min_rate, 2)
    })
    
    print(f"\n{col.upper()}:")
    print(f"  Best: {max_cat} ({max_rate:.2f}%)")
    print(f"  Worst: {min_cat} ({min_rate:.2f}%)")

print("\n" + "-"*140)
print("CATEGORICAL FEATURES SUMMARY TABLE")
print("-"*140)
categorical_summary_df = pd.DataFrame(categorical_summary)
print(categorical_summary_df.to_string())
```

### Generating analysis summary report

```python
### Generating analysis summary report

print("-"*140)
print("GENERATING ANALYSIS SUMMARY REPORT")
print("-"*140)

os.makedirs("../outputs/reports", exist_ok=True)

with open("../outputs/reports/statistical_summary.csv", "w") as f:
    f.write("Feature,Mean_No,Mean_Yes,Median_No,Median_Yes,Difference_Percent\n")
    for item in summary_data:
        f.write(f"{item['Feature']},{item['Mean (No)']},{item['Mean (Yes)']},"
                f"{item['Median (No)']},{item['Median (Yes)']},{item['Difference %']}\n")

print("Statistical summary saved to: ../outputs/reports/statistical_summary.csv")

print("\n" + "-"*140)
print("NOTEBOOK 03 COMPLETED SUCCESSFULLY")
print("-"*140)
print("\nProceed to Notebook 04 for data visualization.")
```
