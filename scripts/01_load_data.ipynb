## Notebook 01: Data Loading

### Objective
This notebook focuses on loading the raw bank marketing dataset into memory, performing an initial inspection, and verifying data integrity before proceeding with cleaning and analysis.

### Outcomes
- Load CSV data using pandas
- Inspect basic dataset characteristics (shape, info, descriptive statistics)
- Identify data types and missing values
- Save the raw data summary for documentation

### Steps Covered
1. Import necessary libraries
2. Load the dataset
3. Display basic information
4. Check for missing values
5. Display sample rows
6. Generate summary statistics
7. Save data summary to a report file

### Importing libraries and loading the bank dataset

```python
### Importing libraries and loading the bank dataset

import pandas as pd
import numpy as np
import os

RAW_DATA_PATH = "../data/bank-full.csv"
RANDOM_SEED = 42

print("Current working directory:", os.getcwd())
print("-"*140)
print("LOADING DATA")
print("-"*140)

df = pd.read_csv(RAW_DATA_PATH, sep=';')

print(f"\nDataset shape: {df.shape[0]} rows and {df.shape[1]} columns")

print("\n" + "-"*140)
print("FIRST 5 ROWS OF THE DATASET")
print("-"*140)
print(df.head())

print("\n" + "-"*140)
print("LAST 5 ROWS OF THE DATASET")
print("-"*140)
print(df.tail())
```

### Checking data types and missing values

```python
### Checking data types and missing values

print("-"*140)
print("DATA TYPES AND NON-NULL COUNTS")
print("-"*140)
print(df.info())

print("\n" + "-"*140)
print("CHECKING FOR MISSING VALUES")
print("-"*140)

missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values[missing_values > 0] if any(missing_values > 0) else "No missing values found.")
```

### Generating summary statistics

```python
### Generating summary statistics

print("-"*140)
print("SUMMARY STATISTICS FOR NUMERICAL COLUMNS")
print("-"*140)
print(df.describe())

print("\n" + "-"*140)
print("SUMMARY STATISTICS FOR CATEGORICAL COLUMNS")
print("-"*140)
print(df.describe(include=['object']))
```

### Displaying column information and duplicate check

```python
### Displaying column information and duplicate check

print("-"*140)
print("COLUMN INFORMATION")
print("-"*140)

column_info = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes.values,
    'Unique Values': [df[col].nunique() for col in df.columns]
})
print(column_info.to_string())

print("\n" + "-"*140)
print("DUPLICATE CHECK")
print("-"*140)
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

### Saving data summary to report file

```python
### Saving data summary to report file

import os

os.makedirs("../outputs/reports", exist_ok=True)

with open("../outputs/reports/data_quality_report.txt", "w") as f:
    f.write("-"*140 + "\n")
    f.write("BANK MARKETING DATASET - DATA QUALITY REPORT\n")
    f.write("-"*140 + "\n\n")
    
    f.write(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns\n\n")
    
    f.write("COLUMN INFORMATION:\n")
    f.write("-"*40 + "\n")
    for col in df.columns:
        f.write(f"  {col}: {df[col].dtype} (unique: {df[col].nunique()})\n")
    
    missing_values = df.isnull().sum()
    f.write("\nMISSING VALUES SUMMARY:\n")
    f.write("-"*40 + "\n")
    if missing_values.sum() == 0:
        f.write("  No missing values found in the dataset.\n")
    else:
        for col, val in missing_values[missing_values > 0].items():
            f.write(f"  {col}: {val} missing\n")
    
    f.write(f"\nDUPLICATE ROWS: {duplicate_count}\n")
    
    if 'y' in df.columns:
        f.write("\nTARGET VARIABLE DISTRIBUTION:\n")
        f.write("-"*40 + "\n")
        target_counts = df['y'].value_counts()
        for label, count in target_counts.items():
            f.write(f"  {label}: {count} ({count/len(df)*100:.2f}%)\n")

print("Data quality report saved to: ../outputs/reports/data_quality_report.txt")
print("-"*140)
print("NOTEBOOK 01 COMPLETED SUCCESSFULLY")
print("-"*140)
```
